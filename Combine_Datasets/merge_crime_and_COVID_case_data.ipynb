# Loading Crime Only dataset (original from DC Crime Data)

```python
df_pre_covid_crime_only = pd.read_csv(os.path.join(data_dir, "Merged", "crime_pre-covid_time_data.csv"))
df_covid_crime_only = pd.read_csv(os.path.join(data_dir, "Merged", "crime_covid_time_data.csv"))

df_pre_covid_crime_only['START_DATE'] = pd.to_datetime(df_pre_covid_crime_only['START_DATE'])
df_covid_crime_only['START_DATE'] = pd.to_datetime(df_covid_crime_only['START_DATE'])

df_pre_covid_crime_only = df_pre_covid_crime_only.set_index(df_pre_covid_crime_only['START_DATE'])
df_covid_crime_only = df_covid_crime_only.set_index(df_covid_crime_only['START_DATE'])

del df_pre_covid_crime_only["START_DATE.1"]
del df_covid_crime_only["START_DATE.1"]

df_pre_covid_crime_only['date'] = pd.to_datetime(df_pre_covid_crime_only['START_DATE'].dt.date)
df_covid_crime_only['date'] = pd.to_datetime(df_covid_crime_only['START_DATE'].dt.date)


print("df_pre_covid_crime_only.shape:", df_pre_covid_crime_only.shape)
print("df_covid_crime_only.shape:", df_covid_crime_only.shape)
```

# Adding COVID-cases data

```python
df_covid_cases_data_only = pd.read_csv(os.path.join(data_dir, "Merged", "covid_cases_district-of-columbia-history.csv"))

df_covid_cases_data_only = df_covid_cases_data_only[(df_covid_cases_data_only["state"])=="District of Columbia"]
df_covid_cases_data_only["covid_deaths"] = df_covid_cases_data_only['deaths']
df_covid_cases_data_only['covid_daily_case_count'] = df_covid_cases_data_only['cases']

del df_covid_cases_data_only["state"]
del df_covid_cases_data_only["deaths"]
del df_covid_cases_data_only["cases"]
del df_covid_cases_data_only["fips"]
```

```python
df_covid_cases_data_only
```

```python
# fix the column types and sort by date
```

```python
df_covid_cases_data_only['index_date'] = pd.to_datetime(df_covid_cases_data_only['date'])
#df_covid_cases_data['Case_Date'] = df_covid_cases_data_only['date']
df_covid_cases_data_only = df_covid_cases_data_only.set_index(df_covid_cases_data_only['index_date']) #.dt.strftime('%Y-%m-%d')) # should it be END_DATE or what?
df_covid_cases_data_only = df_covid_cases_data_only.sort_index()

df_covid_cases_data_only = df_covid_cases_data_only[['date', 'covid_daily_case_count', 'covid_deaths']]
df_covid_cases_data_only['date'] = pd.to_datetime(df_covid_cases_data_only['date'])
df_covid_cases_data_only = df_covid_cases_data_only.rename(columns={"positive":"covid_daily_case_count"})
```

# Example for reading in merged dataset that includes Crime and COVID case data

```python
df_pre_covid_crime_and_covid_cases = pd.read_csv(os.path.join(data_dir, "Merged", "pre_covid_crime_and_covid_cases.csv"))
df_covid_crime_and_cases = pd.read_csv(os.path.join(data_dir, "Merged", "covid_crime_and_cases.csv"))

df_pre_covid_crime_and_covid_cases['REPORT_DAT'] = pd.to_datetime(df_pre_covid_crime_and_covid_cases['REPORT_DAT'])
df_pre_covid_crime_and_covid_cases['START_DATE'] = pd.to_datetime(df_pre_covid_crime_and_covid_cases['START_DATE'])
df_pre_covid_crime_and_covid_cases['END_DATE'] = pd.to_datetime(df_pre_covid_crime_and_covid_cases['END_DATE'])

df_covid_crime_and_cases['REPORT_DAT'] = pd.to_datetime(df_covid_crime_and_cases['REPORT_DAT'])
df_covid_crime_and_cases['START_DATE'] = pd.to_datetime(df_covid_crime_and_cases['START_DATE'])
df_covid_crime_and_cases['END_DATE'] = pd.to_datetime(df_covid_crime_and_cases['END_DATE'])
```

```python
df_pre_covid_crime_and_covid_cases = df_pre_covid_crime_and_covid_cases.dropna()
df_covid_crime_and_cases = df_covid_crime_and_cases.dropna()
```
